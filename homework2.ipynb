```python
import matplotlib.pyplot as plt
import numpy as np
import cv2
```

```python
image=  cv2.imread("ev.jpg")

image = cv2.cvtColor(image,cv2.COLOR_BGR2RGB)
original_shape = image.shape
print(image.shape)
```

```python
plt.imshow(image)
```

```python
from sklearn.cluster import KMeans

x = image.reshape(-1, 3) # (426400, 3)

segmentation = KMeans(n_clusters=4, random_state=42)
kmeans = segmentation.fit(x)
# print(kmeans.cluster_centers_)
# print(kmeans.labels_.shape) # (426400,)
segmented_image = kmeans.cluster_centers_[kmeans.labels_]
segmented_image = segmented_image.reshape(image.shape)


cv2.imwrite("segmented1.jpg", segmented_image)
```
